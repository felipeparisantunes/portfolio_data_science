# Panorama do COVID-19 no Brasil

COVID-19 é uma doença infecciosa causada por recém-descoberto coronavírus.

Transmitido principalmente por meio de gotículas provenientes de tosses ou espirros de pessoas infectadas, a gravidade dos sintomas varia muito de pessoa para pessoa.

Fato é, não se sabe muita coisa a respeito do COVID-19. Estudos estão sendo realizados no mundo todo, porém os resultados ainda não são conclusivos e definitivos.

<center><img alt="Colaboratory logo" width="50%" src="https://raw.githubusercontent.com/carlosfab/dsnp2/master/img/covid19.jpg"></center>

Até o presente momento, observa-se que cerca de 80% dos casos confirmados são assintomáticos e rápidos. A maioria das pessoas que se encaixam nesse grupo, se recupera sem nenhuma sequela.

No entanto, 15% das pessoas terão infecções graves e precisarão de oxigênio. O restante das pessoas, que representam 5%, serão classificadas como infecções muito graves e precisarão de ventilação assistida, por meio de respiradores mecânicos em ambiente hospitalar.

Com o objetivo de elevar a consciência situacional a respeito do COVID-19 no Brasil, irei realizar uma análise sobre os dados públicos da doença.

### Localizaçãos dos Dados Brutos 

* Os dados brutos estão disponíveis no link [https://raw.githubusercontent.com/owid/covid-19-data/master/public/data/owid-covid-data.csv]. Usaremos o formato `csv` neste projeto.
    

```
# bibliotecas necessárias
import pandas as pd
import matplotlib.pyplot as plt

# Dataset COVID-19 para um DataFrame
df = pd.read_csv('https://raw.githubusercontent.com/owid/covid-19-data/master/public/data/owid-covid-data.csv')
```

```
df.head()
```

## Análise Exploratória dos Dados

*Na Análise Exploratória Inicial vamos começar verificando uma amostra os dados, verificando os tipos de colunas, tamanho do Dataframe, Valores Ausentes e entre outros.*

*Está Análise nos ajudará a compreender melhor os dados contidos neste Dataframe, também será possível realizar uma limpeza de dados não úteis, e por consequência deixando-o apto para extração de Insights.*


```
# Tamanho Total do DF
print('Este Dataframe possui {} Linhas e {} colunas'.format(df.shape[0], df.shape[1]))
```

```
# Tipos das Variáveis
#print(df.dtypes)

# Ajustando o Campo de Data para Tipo Datetime, para facilitar impressões de gráficos usando Tempo
df.date = pd.to_datetime(df.date)

# Conferir alteração do Tipo de Dados da Data
print(df.dtypes)
```

```
# Identificar Dados Ausentes 
(df.isnull().sum() / df.shape[0] * 100).sort_values(ascending=False)
```

*Observamos que existem colunas com muitos dados ausentes, mais de 95%, ou seja, não contruibuindo em nada em nosso Dataset. Para esta análise não será necessário remover estas colunas, porém, caso fosse realizado um modelo para Machine Learning seria altamente aconselhável a remoção ou preenchimento delas.*

```
# Data máxima do Dataframe
print('Esta pesquisa contém dados até: {}'.format(df.date.max()))
```

```
# Realizando uma Soma dos Novos Casos por Continente, podemos ver que até a Data deste Dataframe a Europa liderava o Número de Casos de Covid-19.
df.groupby(['continent']).new_cases.sum().sort_values(ascending=False)
```

```
# Agora verificando o número casos por País
df.loc[df.date == '2022-02-13'].sort_values(by='total_cases', ascending=False)[0:13]
```

Observamos que considerando o Planeta, em números de casos, temos:

1º - Estados Unidos com 77.739.880 casos.

2º - India com 42.665.534 casos.

3º - Brasil com 27.492.904 casos.

4º - Franca com 21.761.364 casos.

```
# Vamor Verificar também o número de mortes por País
df.loc[df.date == '2022-02-13'].sort_values(by='total_deaths', ascending=False)[0:13]
```

Para o Total de Mortes entre os países, o Ranking fica desta forma:

1º - Estados Unidos com 919.696 mortes.

2º - Brasil com 638.673 mortes. 

3º - India com 509.011 mortes.

4º - Russia com 333.421 mortes.

### **Vamos realizar Amostras Gráficas para melhor visualização das informações acima.**

```
# Gráfico com Total de Mortes por Localização
top_5_total_deaths = df.loc[df.date == '2022-02-13', ['location', 'total_deaths']].sort_values(by="total_deaths", ascending=False)[1:6]

fig, ax = plt.subplots(figsize=(15,10))

ax.bar('location', 'total_deaths', data=top_5_total_deaths)

plt.tight_layout()
```

```
# Gráfico da evolução do total de casos de COVID-19 no mundo.
df.loc[df.location == "World", "total_deaths"].reset_index(drop=True).plot();
```

### Análise Exploratória para o Brasil

*Nesta etapa abordamos uma visão focada no Brasil, iremos identificar o avanço da Covid-19 em nosso país.*


```
# Primeiro criamos um Dataframe filtrando somente os dados inerentes ao Brasil
df_brasil = df.loc[df.location == "Brazil"].copy()

df_world = df.loc[df.location == "World"].copy()

#Analisamos as 5 primeiras entradas
df_brasil.head()
```

```
#Primeiro Caso registrado no Brasil
print('O Primeiro caso registrado no Brasil foi em: {}.'.format(df_brasil.date.min()))
```

```
#Primeira morte registrada no Brasil
df_brasil.loc[df_brasil['new_deaths'] > 0.0].head(1)

print('A primeira morte Registrada no Brasil foi em {}.'.format(df_brasil.date.loc[20751]))
```

Agora vamos demonstrar quantos dias se passaram entre o primeiro caso e a primeira morte no Brasil.

```
from datetime import date
delta = df_brasil.date.loc[20751] - df_brasil.date.loc[20731]
print('Se passaram {} dias entre o primeiro caso e a primeira morte no Brasil.'.format(delta.days))
```

```
# Verificando Total de Mortes no Brasil até o momento
df_brasil.tail()
df_world.tail()

print('Até o dia 14/02/2022 no Brasil foram registradas {} mortes pela Covid-19, enquanto no Mundo foram {} mortes.'.format(df_brasil.total_deaths.loc[21450], df_world.total_deaths.loc[159836]))
```

```
# Verificando o dia com mais casos no Brasil
max_new_cases_brazil = df_brasil[['date', 'new_cases']].sort_values(by='new_cases', ascending=False).head(1)
print(max_new_cases_brazil.loc[21439])


```

```
# Verificando o dia com mais mortes no Brasil
max_new_deaths_brazil = df_brasil[['date', 'new_deaths']].sort_values(by='new_deaths', ascending=False).head(1)
print(max_new_deaths_brazil.loc[21138])
```

Observamos que o dia com mais Casos foi em 03/02/2022 com 287.149 casos.

O Dia com mais mortes no Brasil foi em 08/04/2021 com 4.148 mortes.

```
# Verificando o Números de Casos e Mortes sobre a População Brasileira
population_brazil = df_brasil.population.loc[21450]
print('A População do Brasil é {}.'.format(population_brazil))

total_mortes_brasil = df_brasil.total_deaths.loc[21450]
print('Até o momento foram registradas {} mortes pela Covid-19 no Brasil.'.format(total_mortes_brasil))

total_casos_brasil = df_brasil.total_cases.loc[21450]
print('Até o momento foram registrados {} casos de Covid-19 no Brasil.'.format(total_casos_brasil))

# Percentual de casos sobre a população
print('{}% das pessoas do brasil foram infectadas pelo vírus da Covid.'.format(round(total_casos_brasil / population_brazil * 100),2))


# Percentual de mortes sobre a população
print('{}% das pessoas do brasil perderam a vida pelo vírus da Covid.'.format(round(total_mortes_brasil / population_brazil * 100,2)))
```

```
# Representatividade do Brasil perante o Planeta

perc_brasil = round(df_brasil.new_deaths.sum() / df_world.total_deaths.loc[159836] * 100, 2)

print('O Brasil até a Data de 14/02/2022 representava {}% das mortes por Covid-19 do planeta.'.format(perc_brasil))
```

Abaixo representações gráficas inerentes aos Casos e Mortes registradas por Covid-19 no Brasil.

```
x = list(range(720))
z = df_brasil['total_cases']
z2 = df_brasil['total_deaths']

fig, ((ax1, ax2), (ax3, ax4)) = plt.subplots(nrows=2, ncols=2, figsize=(9, 7))
ax1.set_title('Casos no Brasil')
ax1.set_xlabel('Dias')
ax1.set_ylabel('Casos')
ax1.plot(x, z)

ax2.set_title('Casos no Brasil (Escala logarítmica)')
ax2.set_xlabel('Dias')
ax2.set_ylabel('Casos')
ax2.set_yscale('log')
ax2.plot(x, z)

ax3.set_title('Mortes no Brasil')
ax3.set_xlabel('Dias')
ax3.set_ylabel('Mortes')
ax3.plot(x, z2)

ax4.set_title('Mortes no Brasil (Escala logarítmica')
ax4.set_xlabel('Dias')
ax4.set_ylabel('Mortes')
ax4.set_yscale('log')
ax4.plot(x, z2)
plt.tight_layout()
```
